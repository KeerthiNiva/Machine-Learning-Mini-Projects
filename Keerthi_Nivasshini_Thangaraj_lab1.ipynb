```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
#Reading Dataframe
data = pd.read_csv('/content/Lab1_kc_house_data.csv')
```

### Creating model for living space vs price

```python
#Setting x and y parameters
x = data[['sqft_living']]
y = data['price']
```

```python
#Creating model to fit y and x
model = sm.OLS(y, x).fit()
```

```python
#Prediction data
prediction = model.predict(pd.DataFrame({'sqft_living': [1000]}))
print(prediction)
```

```python
import matplotlib.pyplot as plt
# prediction for a range of square footage values
predicted_values = model.predict(x)

# plotting the actual data and the predicted data
plt.scatter(x, y, color='red', label='Actual Data')
plt.plot(x, predicted_values, color='blue', label='Predicted Data')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Actual vs Predicted Price of Houses')
plt.legend()
plt.show()
```
