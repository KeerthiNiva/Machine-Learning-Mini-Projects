**1.Import pandas, numpy, and matplotlib, as well as the t-SNE and PCA models from scikit-learn**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

**2.Load the Wine dataset in datafram without header and display the first five rows of data.**

```python
ds = pd.read_csv('/content/wine.data', header=None)
print(ds.head())
```

**3.The first column contains the labels; extract this column as labels and remove it from the dataset.**

```python
l = ds.iloc[:, 0]
d = ds.iloc[:, 1:]
```

**4.Execute PCA to reduce the dataset to the first six components.**

```python
pca = PCA(n_components=6)
pca_data = pca.fit_transform(d)
```

**5.Determine the amount of variance within the data described by these six components.**

```python
print("Explained variance of the first six components:", np.sum(pca.explained_variance_ratio_))
```

**6.A t-SNE model using a specified random state and a verbose value of 1**

```python
tsne = TSNE(n_components=2, random_state=42, verbose=1)
```

**7.Fit the data to the t-SNE model.**

```python
t_d = tsne.fit_transform(pca_data)
```

**8.Confirm that the shape of the t-SNE fitted data is two-dimensional**

```python
print("Shape of t-SNE fitted data:", t_d.shape)
```

**9.A scatter plot of the two-dimensional data (first and second features).**

```python
plt.scatter(t_d[:, 0], t_d[:, 1])
plt.title("t-SNE Scatter Plot")
plt.xlabel("First Feature")
plt.ylabel("Second Feature")
plt.show()
```

**10. A secondary scatter plot of the two-dimensional data with the class labels applied to visualize any clustering that may be present**


```python
plt.scatter(t_d[:, 0], t_d[:, 1], c=l)
plt.title("t-SNE Scatter Plot with Class Labels")
plt.xlabel("First Feature")
plt.ylabel("Second Feature")
plt.show()
```

**11.A loop that iterates through the perplexity values (1, 5, 20, 30, 80, 160, 320). For each loop, generate a t-SNE model with the corresponding.**

```python
perplexities = [1, 5, 20, 30, 80, 160, 320]

for p in perplexities:
    tsne = TSNE(n_components=2, perplexity=30, random_state=42, verbose=1)
    t_d = tsne.fit_transform(pca_data)
    plt.scatter(t_d[:, 0], t_d[:, 1], c=l)
    plt.title("t-SNE Scatter Plot with Perplexity = " + str(p))
    plt.xlabel("First Feature")
    plt.ylabel("Second Feature")
    plt.show()
```
